### Which mode of call is giving more result

```python
df['contact'].value_counts()
```

```python
cellular_output = df[(df['contact'] =='cellular') & (df['y'] == 'yes')]['y']
cellular = len(cellular_output)
print("Cellular=", cellular)
```

```python
telephone_output = df[(df['contact'] == 'telephone') & (df['y'] == 'yes')]['y']
telephone = len(telephone_output)
print("Telephone=", telephone)
```

```python
unknown_output =  df[(df['contact'] == 'unknown') & (df['y'] == 'yes')]['y']
unknown = len(unknown_output)
print("Unknown=", unknown)
```
