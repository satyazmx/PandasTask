### Find how many males and females onboarded

```python
male_count = df[df['Sex'] == 'male']['Sex'].count()
print("Male count =", male_count)
```

```python
female_count = df[df['Sex'] == 'female']['Sex'].count()
print("Female count =", female_count)
```

### Find number of passangers survived

```python
survived = df[df['Survived'] == 1]['Survived'].count()
print("survived count =", survived)
```

### Find number of casualties

```python
casualty = df[df['Survived'] == 0]['Survived'].count()
print("casualty count =", casualty)
```

### Find eldest passanger's name 

```python
df[df['Age'] == df['Age'].max()]['Name']
```

### number of passangers in each class

```python
df['Pclass'].unique()
```

```python
first_class = df[df['Pclass'] == 1]['Pclass'].count()
print("First Class count =", first_class)
```

```python
second_class = df[df['Pclass'] == 2]['Pclass'].count()
print("Second Class count =", second_class)
```

```python
total = df['Pclass'].count()
third = total - (first_class + second_class)
print("Third Class count =", third)
```

### NUmmber of passangers whose name starts with s

```python
starts_with_s = df[df['Name'].str.startswith('S')]['Name'].count()
print("starts_with_s count =", starts_with_s)
```

### Create new column summation of Sibsp & Parch

### Passengers having age less than 25

```python
age = df[df['Age'] <= 25]['Age'].count()
print("Age less than 25 count =", age)
```

### Passangers died whose age is lesser than 40 Yrs

```python
output = len(df[(df['Survived'] == 0) & (df['Age'] < 40)])
print("Died passanger with age esser than 40 = ",output)
```

### Extract numeric andalphabet seprate of Cabin

```python
df['Cabin_Letter'] = df['Cabin'].str.extract('([A-Za-z]+)')
df['Cabin_Number'] = df['Cabin'].str.replace('([A-Za-z]+)','')
```
